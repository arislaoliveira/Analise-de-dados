```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Dados6.csv')
```

```python
df['altura'] = pd.to_numeric(df['altura'], errors='coerce')
df['peso'] = pd.to_numeric(df['peso'], errors='coerce')
df['idade'] = pd.to_numeric(df['idade'], errors='coerce')
df['distancia'] = pd.to_numeric(df['distancia'], errors='coerce')
df['moradores'] = pd.to_numeric(df['moradores'], errors='coerce')
df.drop(columns=['Unnamed: 7'], inplace = True)
```

```python
df.sort_values(by='altura', ascending=False)
```

```python
df.sort_values(by='peso', ascending=False)
```

```python
df['IMC'] = df['peso'] / (df['altura'] ** 2)
df.sort_values(by='IMC', ascending=False)
```

```python
def classificar_imc(imc):
    if imc < 18.5:
        return 'Abaixo do peso'
    elif 18.5 <= imc < 25:
        return 'Peso normal'
    elif 25 <= imc < 30:
        return 'Sobrepeso'
    elif 30 <= imc < 35:
        return 'Obesidade Grau I'
    elif 35 <= imc < 40:
        return 'Obesidade Grau II (severa)'
    else:
        return 'Obesidade Grau III (mórbida)'

df['Classificação IMC'] = df['IMC'].apply(classificar_imc)

df.sort_values(by='IMC', ascending=False)
```
