```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Dados6.csv')
```

```python
df_limpo = df.dropna()

df_limpo
```

```python
media_peso = df['peso'].mean()

df['peso'] = df['peso'].fillna(media_peso)
```

```python
df = df.drop('Unnamed: 7', axis=1)
print(df)
```

```python
df_sem_duplicados = df.drop_duplicates()
display(df_sem_duplicados)
```

```python
moradores_max = df['moradores'].max()
moradores_min = df['moradores'].min()

df['moradores_normal'] = (df['moradores'] - moradores_min) / (moradores_max - moradores_min)
```

```python
df = df.drop(56)

df['peso_inteiro'] = df['peso'].astype(int)
```

```python
df['moradores_inteiro'] = df['moradores_normal'].astype(int)
print(df)
```

```python
df.to_csv('dados_limpos.csv', index=False)
```
