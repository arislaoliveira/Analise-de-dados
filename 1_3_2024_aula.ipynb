```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = {'Mês': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho'],
         'Vendas': [50, 45, 60, 55, 70, 65]}
df = pd.DataFrame(dados)
print(df.info())
```

```python
descricao = df['Vendas'].describe()
print(descricao)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Mês'], df['Vendas'], color='skyblue')
plt.title('Vendas Mensais')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.show()
```
